```python
researcher = Agent(
    role="Senior Research Analyst",
    goal="Investigate and report on the latest advancements in OTT industry in 2024.",
    backstory="""You work at a prominent Media House.
    Your expertise lies in sourcing and analyzing information on OTT industry. You excel at breaking down complex data and presenting it in an accessible and insightful manner.""",
    verbose=False,
    allow_delegation=False,
    tools=[search_tool],
    llm=ChatGroq(model_name="llama3-8b-8192", temperature=0.7)
)

writer = Agent(
    role="Expert Technology Writer",
    goal="Craft concise and informative articles summarizing the latest advancements in OTT Industry in 2024.",
    backstory="""You are a well-respected content strategist with a knack for creating engaging and informative articles.
    Your expertise lies in transforming complex concepts into clear, compelling narratives that are easily understood by a broad audience.""",
    verbose=True,
    allow_delegation=True,
    llm=ChatGroq(model_name="llama3-8b-8192", temperature=0.7)
)
```

```python
task1 = Task(
    description="Investigate the most recent breakthroughs in OTT Industry.",
    expected_output="A detailed summary of the latest innovations in OTT industry.",
    agent=researcher
)

task2 = Task(
    description="Compose a concise and informative article highlighting the latest innovations in OTT industry.",
    expected_output="An engaging and well-structured article on the recent advancements in OTT industry.",
    agent=writer
)
```

```python
crew = Crew(
    agents = [researcher, writer],
    tasks = [task1, task2],
    verbose = 1
)
```

```python
final = crew.kickoff()
```
